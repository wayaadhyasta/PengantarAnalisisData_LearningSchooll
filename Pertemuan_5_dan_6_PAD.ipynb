````python
missing = ['na', '--']
df = pd.read_csv('/content/Apartemen.csv', na_values = missing)
````

````python
#Baca data kolom st milik
df['St_Milik']
````

````python
df['St_Milik'].isnull()
````

````python

cnt = 0

for row in df['St_Milik']:
  try:
    df['St_Milik'].isnull()
    int(row) #Mendeteksi angka pada kolom st milik
    df.loc[cnt, 'St_Milik'] = np.nan #apabila int maka akan disimpan pada variabel np.nan
  except ValueError:
    pass
  cnt+=1
````

````python
df['St_Milik'].isnull() #isnull untuk memastikan data ada missing value atau tidak
````

# Scalling


````python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('/content/Apartemen_numerik.csv')
df
````

````python
#Menampilkan statistik deskriptif
df.describe()
````

````python
df.loc[2, 'Jum_kamar'] = 100
df
print(df.shape)
df['Jum_kamar']
````

````python
#fitur scalling
fitur_minmax = df.copy()
Kol_minmax = ['Jum_kamar']
features = fitur_minmax[Kol_minmax]
min_max= MinMaxScaler()
kamar_minmax = min_max.fit_transform(features.values)
kamar_minmax
````

````python
for col in df.columns.values:
  features = df[[col]]
  features_minmax = min_max.fit_transform(features)
  df[col] = features_minmax
df
````
